```python
import pandas as pd
import json
import os
from PIL import Image
import numpy as np
```

```python
with open('img_emotion_training_data(기쁨)2.json', 'r', encoding='utf-8') as file:
    data = json.load(file)
```

```python
rows = []
```

```python
# 이미지 처리 및 데이터프레임 생성 과정
for item in data:
    filename = item["filename"]
    gender = item["gender"]
    age = item["age"]
    faceExp_uploader = item["faceExp_uploader"]
    bg_uploader = item["bg_uploader"]
    
    for annot_name in ['annot_A', 'annot_B', 'annot_C']:
        annot = item[annot_name]
        faceExp = annot["faceExp"]
        bg = annot["bg"]
        
        # 이미지 경로
        image_path = f'imgs/{filename}'

        # 이미지 파일이 존재하는지 확인
        if os.path.isfile(image_path):
            # 이미지를 열고 좌표에 따라 자르고 저장
            img = Image.open(image_path)
            bbox = (
                int(annot["boxes"]["minX"]),
                int(annot["boxes"]["minY"]),
                int(annot["boxes"]["maxX"]),
                int(annot["boxes"]["maxY"])
            )
            cropped_img = img.crop(bbox)

            # 이미지 저장 경로 (imgs 폴더 내 cropped 폴더에 저장)
            save_path = f'imgs/cropped/{filename}_{annot_name}.jpg'
            cropped_img.save(save_path)
            
            rgb_array = np.array(cropped_img)

            # 데이터를 리스트에 추가
            rows.append([rgb_array, filename, gender, age, faceExp_uploader, bg_uploader, faceExp, bg])
        else:
            print(f"Image not found for filename: {filename}")

# 데이터프레임 생성
columns = ['image', 'filename', 'gender', 'age', 'faceExp_uploader', 'bg_uploader', 'faceExp', 'bg']
df = pd.DataFrame(rows, columns=columns)
```
